```python
def compare_common_cards(deck1, deck2):
    commonCards = 0
    for card in deck1['Card'].values:
        if card in deck2['Card'].values:
            #Both decks have at least one copy
            count1 = deck1.loc[(deck1['Card'] == card)]['Quantity'].iloc[0]
            count2 = deck2.loc[(deck2['Card'] == card)]['Quantity'].iloc[0]
            if count1 == count2:
                commonCards += count1
            elif count1 < count2:
                commonCards += count1
            elif count1 > count2:
                commonCards += count2
    return commonCards
```
